```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample data
data = {
    "Dataset": ["DS1", "DS2", "DS3", "DS4", "DS5", "DS6"] * 3,
    "Weight Error Percentage": [
        0.5,
        0.4,
        0.8,
        0.3,
        0.25,
        0.6,
        0.7,
        0.6,
        1.0,
        0.4,
        0.5,
        0.8,
        0.6,
        0.5,
        0.9,
        0.35,
        0.3,
        0.7,
    ],
    "Algorithm": ["Exact"] * 6 + ["FMST"] * 6 + ["Proposed"] * 6,
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x="Dataset", y="Weight Error Percentage", hue="Algorithm", data=df)
plt.title("Comparison of Weight Error Percentage Across Different Algorithms")
plt.xlabel("Dataset")
plt.ylabel("Weight Error Percentage")
plt.legend(title="Algorithm")
plt.show()
```
